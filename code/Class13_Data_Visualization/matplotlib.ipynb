```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, y)  # Plot some data on the axes.
ax.set_title('Simple Plot - OO Interface')  # Add a title to the axes.
ax.set_xlabel('X Axis')  # Add an x-label to the axes.
ax.set_ylabel('Y Axis')  # Add a y-label to the axes.
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

fig = plt.figure()  # Create a figure 
ax = fig.add_subplot()  # Add a subplot
ax.plot(x, y)  # Plot some data on the axes.
ax.set_title('Simple Plot - OO Interface')  # Add a title to the axes.
ax.set_xlabel('X Axis')  # Add an x-label to the axes.
ax.set_ylabel('Y Axis')  # Add a y-label to the axes.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = range(1, 5)
y1 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]

# First subplot
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y1)
plt.title('First Subplot')

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Second Subplot')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
x = range(1, 5)
y1 = [1, 2, 3, 4]
y2 = [4, 3, 2, 1]

# Create figure and axes objects
fig, axs = plt.subplots(1, 2)  # (rows, columns)

# First subplot
axs[0].plot(x, y1)
axs[0].set_title('First Subplot')

# Second subplot
axs[1].plot(x, y2)
axs[1].set_title('Second Subplot')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Create a plot using Pyplot
plt.plot(x, y)

# Grab the current Axes and Figure using Pyplot, then modify using OO
ax = plt.gca()  # Get current Axes
fig = plt.gcf()  # Get current Figure

# Now, use OO methods
ax.set_title('Plot Title Using OO')
ax.set_xlabel('X Label Using OO')
ax.set_ylabel('Y Label Using OO')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Start with OO
fig, ax = plt.subplots()
ax.plot(x, y)

# Use Pyplot to modify, but this affects the current plot
plt.title('Plot Title Using Pyplot')
plt.xlabel('X Label Using Pyplot')
plt.ylabel('Y Label Using Pyplot')

plt.show()



```
